# About Dataset

<b>Context</b>

The growth of supermarkets in most populated cities are increasing and market competitions are also high. The dataset is one of the historical sales of supermarket company which has recorded in 3 different branches for 3 months data. Predictive data analytics methods are easy to apply with this dataset.

<b>Attribute information</b>

* Invoice id: Computer generated sales slip invoice identification number.
* Branch: Branch of supercenter (3 branches are available identified by A, B and C).
* City: Location of supercenters.
* Customer type: Type of customers, recorded by Members for customers using member card and Normal for without member card.
* Gender: Gender type of customer.
* Product line: General item categorization groups - Electronic accessories, Fashion accessories, Food and beverages, Health and beauty, Home and lifestyle, Sports and travel.
* Unit price: Price of each product in $.
* Quantity: Number of products purchased by customer.
* Tax: 5% tax fee for customer buying.
* Total: Total price including tax.
* Date: Date of purchase (Record available from January 2019 to March 2019).
* Time: Purchase time (10am to 9pm).
* Payment: Payment used by customer for purchase (3 methods are available â€“ Cash, Credit card and Ewallet).
* COGS: Cost of goods sold.
* Gross margin percentage: Gross margin percentage.
* Gross income: Gross income.
* Rating: Customer stratification rating on their overall shopping experience (On a scale of 1 to 10).

# Data visualisation

## Total price distribution

```python
sns.set_palette('Spectral')

sns.histplot(data=data, x='Total', kde=True)
plt.title('Total price distribution')

plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(1, 3, figsize=(15, 5))

sns.kdeplot(data=data, x='Total', clip=[0, data['Total'].max()], hue='City', fill=True, ax=ax[0])
ax[0].set_title('Total price distribution by cities')

sns.kdeplot(data=data, x='Total', clip=[0, data['Total'].max()], hue='Customer type', fill=True, ax=ax[1])
ax[1].set_title('Total price distribution by Customer types')

sns.kdeplot(data=data, x='Total', clip=[0, data['Total'].max()], hue='Gender', fill=True, ax=ax[2])
ax[2].set_title('Total price distribution by Gender')

plt.tight_layout()
plt.show()
```

## Product line

```python
plt.figure(figsize=(14, 7))
sns.countplot(data=data, x='Product line', hue='Gender')

plt.title('Product line buyers by gender')

plt.tight_layout()
plt.show()
```

```python
product_line_amount = data.groupby('Product line')['Total'].sum()
product_line_amount
```

```python
plt.pie(product_line_amount, labels=product_line_amount.index, autopct='%.2f%%')
plt.title('Product line expences')
plt.tight_layout()
```

## Rating

```python
sns.histplot(data=data, x='Rating', kde=True)
plt.title('Rating distribution')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(14, 7))

sns.kdeplot(data=data, x='Rating', hue='City', fill=True, clip=[data['Rating'].min(), data['Rating'].max()])
plt.title('Rating distribution by cities')

plt.tight_layout()
plt.show()
```

## Gross income

```python
sns.histplot(data=data, x='gross income', kde=True)
plt.title('Grocery income distribution')

plt.tight_layout()
plt.show()
```

```python
plt.pie(data.groupby('Product line')['gross income'].sum(), labels=data.groupby('Product line')['Total'].sum().index, autopct='%.2f%%')
plt.title('Gross income by product line')
plt.tight_layout()
```
