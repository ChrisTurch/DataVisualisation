# О датасете
This data set consists of the marks secured by the students in various subjects.

* gender : sex of students -> (Male/female);
* race/ethnicity : ethnicity of students -> (Group A, B,C, D,E);
* parental level of education : parents' final education ->(bachelor's degree,some college,master's degree,associate's degree,- high school);
* lunch : having lunch before test (standard or free/reduced);
* test preparation course : complete or not complete before test;
* math score;
* reading score;
* writing score;

# Импорт библиотек

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings("ignore", "use_inf_as_na")
```

# Чтение данных и изучение свойств

```python
data = pd.read_csv('study_performance.csv')
data.head()
```

```python
data.info()
```

```python
data.describe()
```

Создадим дополнительные колонки: 
* total score - сумма набранных баллов
* average - средний балл по трём предметам 

```python
data['total score'] = data['math_score'] + data['reading_score'] + data['writing_score']
data['average'] = data['total score']/3
data.head()
```

```python
data.describe()
```

# Визуализация данных

```python
sns.set_palette('Spectral')

fig, ax = plt.subplots(3, 2, figsize=(14, 14))

sns.countplot(x='gender', data=data, ax=ax[0, 0])
ax[0, 0].set_title('Распределение по полу')

sns.countplot(x='race_ethnicity', data=data, ax=ax[0, 1])
ax[0, 1].set_title('Распределение по расе')

sns.countplot(x='parental_level_of_education', data=data, ax=ax[1, 0])
ax[1, 0].set_title('Распределение по образованию')
ax[1, 0].tick_params(axis='x', rotation=45)

sns.countplot(x='lunch', data=data, ax=ax[1, 1])
ax[1, 1].set_title('Распределение по обеду')

sns.countplot(x='test_preparation_course', data=data, ax=ax[2, 0])
ax[2, 0].set_title('Распределение по прохождению подготовки')

plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(3, 2, figsize=(14, 14))

sns.histplot(x='math_score', data=data, ax=ax[0, 0])
ax[0, 0].set_title('Баллы за математику')

sns.histplot(x='reading_score', data=data, ax=ax[0, 1])
ax[0, 1].set_title('Баллы за чтение')

sns.histplot(x='writing_score', data=data, ax=ax[1, 0])
ax[1, 0].set_title('Баллы за письмо')

sns.histplot(x='total score', data=data, ax=ax[1, 1])
ax[1, 1].set_title('Сумма баллов')

sns.histplot(x='average', data=data, ax=ax[2, 0])
ax[2, 0].set_title('Средний балл')

plt.tight_layout()
plt.show()
```

```python
sns.pairplot(data, hue = 'gender')
plt.show()
```

```python
fig, ax = plt.subplots(1, 5, figsize=(15, 3))

fig.suptitle('Сравнение баллов при (не)прохождении подготовки')

sns.kdeplot(data, x='math_score', hue='test_preparation_course', ax=ax[0], fill=True)
ax[0].set_title('Математика')

sns.kdeplot(data, x='reading_score', hue='test_preparation_course', ax=ax[1], fill=True)
ax[1].set_title('Чтение')

sns.kdeplot(data, x='writing_score', hue='test_preparation_course', ax=ax[2], fill=True)
ax[2].set_title('Письмо')

sns.kdeplot(data, x='total score', hue='test_preparation_course', ax=ax[3], fill=True)
ax[3].set_title('Общий балл')

sns.kdeplot(data, x='average', hue='test_preparation_course', ax=ax[4], fill=True)
ax[4].set_title('Средний балл')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(14, 7))
sns.kdeplot(data=data, x='average', hue='lunch', fill=True)

plt.tight_layout()
plt.show()
```

Средний балл выше, если тестируемый хорошо покушал.
