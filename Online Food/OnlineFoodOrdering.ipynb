# О датасете
Набор данных содержит информацию, собранную с платформы онлайн-заказа продуктов питания за определенный период времени

<b> Атрибуты: </b>

* Age: Age of the customer. 
* Gender: Gender of the customer. 
* Material Status: Marital status of the customer. 
* Occupation: Occupation of the customer.
* Monthly Income: Monthly income of the customer. 
* Educational Qualifications: Educational qualifications of the customer. 
* Family Size: Number of individuals in the customer's family. 
* Latitude: Latitude of the customer's location.
* Longitude:  Longitude of the customer's location. 
* Pin Code:  Pin code of the customer's location. 
* Output:  Current status of the order (e.g., pending, confirmed, delivered).
* Feedback: Feedback provided by the customer after receiving the order.


## Импорт библиотек

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings("ignore", "use_inf_as_na")
```

## Чтение данных и изучение свойств

```python
data = pd.read_csv('onlinefoods.csv')
data.head()
```

```python
data.info()
```

```python
data.describe()
```

## Визуализация

### Распределение по возрасту, размеру семьи 

```python
sns.set_palette('inferno')

fig, ax = plt.subplots(1, 2, figsize=(14, 6))

sns.histplot(data['Age'], bins=15, kde=True, ax=ax[0])
ax[0].set_title('Распределение по возрасту')

sns.histplot(data['Family size'], bins=6, kde=True, ax=ax[1])
ax[1].set_title('Распределение по размеру семьи')

plt.tight_layout()
plt.show()
```

### Распределение по гендеру, семейному положению, занятости и образованию

```python
fig, ax = plt.subplots(2, 2, figsize=(14, 10))

sns.countplot(x='Gender', data=data, ax=ax[0, 0])
ax[0, 0].set_title('Распределение по полу')

sns.countplot(x='Marital Status', data=data, ax=ax[0, 1])
ax[0, 1].set_title('Распределение по семейному положению')

sns.countplot(x='Occupation', data=data, ax=ax[1, 0])
ax[1, 0].set_title('Распределение по занятости')

sns.countplot(x='Educational Qualifications', data=data, ax=ax[1, 1])
ax[1, 1].set_title('Распределение по образованию')

plt.tight_layout()
plt.show()
```

### Онлайн/офлайн заказы

```python
plt.figure(figsize=(14, 7))
sns.histplot(data=data[data['Output'] == 'Yes'], x='Age', bins=15, label='Онлайн')
sns.histplot(data=data[data['Output'] == 'No'], x='Age', bins=15, label='Офлайн')

plt.title('Количество онлайн/офлайн заказов в разбивке по возрасту')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(14, 7))
sns.kdeplot(data=data[data['Output'] == 'Yes'], x='Age', fill=True, label='Онлайн')
sns.kdeplot(data=data[data['Output'] == 'No'], x='Age', fill=True, label='Офлайн')

plt.title('Соотношение онлайн/офлайн заказов в разбивке по возрасту')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(2, 2, figsize=(14, 10))

sns.countplot(x='Gender', data=data, ax=ax[0, 0], hue='Output')
ax[0, 0].set_title('Oнлайн/офлайн заказы по полу')

sns.countplot(x='Marital Status', data=data, ax=ax[0, 1], hue='Output')
ax[0, 1].set_title('Oнлайн/офлайн заказы по семейному положению')

sns.countplot(x='Occupation', data=data, ax=ax[1, 0], hue='Output')
ax[1, 0].set_title('Oнлайн/офлайн заказы по занятости')

sns.countplot(x='Educational Qualifications', data=data, ax=ax[1, 1], hue='Output')
ax[1, 1].set_title('Oнлайн/офлайн заказы по образованию')

plt.tight_layout()
plt.show()
```

### Отзывы

```python
plt.figure(figsize=(14, 7))
sns.histplot(data=data[data['Feedback'] == 'Positive'], x='Age', bins=15, label='Положительные')
sns.histplot(data=data[data['Feedback'] != 'Positive'], x='Age', bins=15, label='Отрицательные')

plt.title('Отзывы в разбивке по возрасту')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(14, 7))
sns.kdeplot(data=data[data['Feedback'] == 'Positive'], x='Age', fill=True, label='Положительные')
sns.kdeplot(data=data[data['Feedback'] != 'Positive'], x='Age', fill=True, label='Отрицательные')

plt.title('Соотношение отзывов в разбивке по возрасту')
plt.legend()

plt.tight_layout()
plt.show()
```

### Связь возраста и размера семьи

```python
plt.figure(figsize=(14, 7))
sns.scatterplot(data=data, x='Age', y='Family size')

plt.title('Связь возраста и размера семьи')

plt.tight_layout()
plt.show()
```

### Связь возраста и размера семьи по полу и семейному статусу

```python
sns.set_palette('Spectral')

plt.figure(figsize=(14, 7))
sns.scatterplot(data=data, x='Age', y='Family size', hue='Gender', style='Marital Status', s=100)

plt.title('Связь возраста и размера семьи по полу и семейному статусу')
plt.legend()

plt.tight_layout()
plt.show()
```
